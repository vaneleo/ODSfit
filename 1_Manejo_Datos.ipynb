```python
# Define the path to your folder containing CSV files
folder_path = '/content/drive/MyDrive/HackODS/elias/hackathon/datas/01_registro_alumnado_personal/data/basica_y_nms/' # <-- IMPORTANT: Replace with your actual folder path

# Create an empty list to store individual DataFrames
df_list = []

# Check if the folder exists
if not os.path.isdir(folder_path):
    print(f"Error: The folder '{9}' was not found. Please ensure the path is correct.")
else:
    # Iterate over all files in the specified folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            print(f"Reading {file_path}...")
            try:
                df = pd.read_csv(file_path, encoding='latin1')
                df = df.rename(columns={'n_entidad': 'n_entidad', 'c_nom_ent': 'n_entidad'})
                df = df.rename(columns={'insc_t': 'insc_t', 'alumnos': 'insc_t'})
                df = df.rename(columns={'hom_t': 'hom_t', 'hombres': 'hom_t'})
                df = df.rename(columns={'muj_t': 'muj_t', 'mujeres': 'muj_t'})
                df = df.rename(columns={'egre_hom': 'egre_hom', 'egresados_h': 'egre_hom'})
                df = df.rename(columns={'egre_muj': 'egre_muj', 'egresados_m': 'egre_muj'})
                df = df.rename(columns={'egre_tot': 'egre_tot', 'egresados': 'egre_tot'})
                df = df.rename(columns={'docente_h': 'docente_h', 'docentes_h': 'docente_h'})
                df = df.rename(columns={'docente_m': 'docente_m', 'docentes_m': 'docente_m'})
                df = df.rename(columns={'tot_doc': 'tot_doc', 'docentes': 'tot_doc'})
                df=df[['entidad','n_entidad','nivel','insc_t','hom_t','muj_t','egre_hom','egre_muj','egre_tot','docente_h','docente_m','tot_doc']]#pd.read_csv(file_path, encoding='latin1')

                df['nivel'] =df['nivel'].str.strip().str.lower()
                df['n_entidad'] =df['n_entidad'].str.strip().str.lower()
                df['n_entidad'] ="".join(caracter for caracter in df['n_entidad'] if caracter.isalnum())
                df=df.groupby(['entidad','n_entidad','nivel'], as_index=False).sum()

                df_list.append(df)
            except Exception as e:
                print(f"Error reading {filename}: {e}")

    if df_list:
        # Concatenate all DataFrames in the list into a single DataFrame
        combined_df = pd.concat(df_list, ignore_index=True)
        print("\nAll CSV files combined successfully. First 5 rows of the combined DataFrame:")
        print(combined_df.head())
    else:
        print("No CSV files found in the specified folder or an error occurred during reading.")
```

```python
# 2. Leer y concatenar todos los archivos
df_final = pd.concat(df_list, ignore_index=True)
```

```python
df_final.info()
```

```python
df_final=df_final.groupby(['entidad','nivel'], as_index=False).sum()
```

```python
df_final=df_final[['entidad','nivel','insc_t','hom_t','muj_t','egre_hom','egre_muj','egre_tot','docente_h','docente_m','tot_doc']]#pd.read_csv(file_path, encoding='latin1')
```

```python
df_final['porc_aban_hom'] = np.where(df_final['hom_t'] == 0, 0, 1 - df_final['egre_hom'] / df_final['hom_t'])
df_final['porc_aban_muj'] = np.where(df_final['muj_t'] == 0, 0, 1 - df_final['egre_muj'] / df_final['muj_t'])
df_final['porc_aban_tot'] = np.where(df_final['insc_t'] == 0, 0, 1 - df_final['egre_tot'] / df_final['insc_t'])
df_final['porc_doc_hom'] = np.where(df_final['tot_doc'] == 0, 0, df_final['docente_h'] / df_final['tot_doc'])
df_final['porc_doc_muj'] = np.where(df_final['tot_doc'] == 0, 0, df_final['docente_m'] / df_final['tot_doc'])
df_final['porc_ins_hom'] = np.where(df_final['insc_t'] == 0, 0, df_final['hom_t'] / df_final['insc_t'])
df_final['porc_insc_muj'] = np.where(df_final['insc_t'] == 0, 0, df_final['muj_t'] / df_final['insc_t'])
df_final
```

```python
df_1=pd.read_csv('/content/drive/MyDrive/HackODS/elias/hackathon/datas/01_registro_alumnado_personal/data/BASICA_2021-2022.csv', encoding='utf8')
df_1=df_1[['entidad','n_entidad']].drop_duplicates()

df_1['n_entidad'] =df_1['n_entidad'].str.strip().str.lower()
df_1['n_entidad']=df_1['n_entidad'].str.replace(r'[^a-zA-Z0-9\s]', '')
df_1
```

```python
df_final_1=df_final.merge(df_1, on='entidad', how='left')
df_final_1=df_final_1[['entidad','n_entidad',	'nivel','insc_t',	'hom_t','muj_t',	'egre_hom',	'egre_muj',	'egre_tot',
                   'docente_h',	'docente_m',	'tot_doc',	'porc_aban_hom',	'porc_aban_muj',	'porc_aban_tot',
                   'porc_doc_hom',	'porc_doc_muj',	'porc_ins_hom',	'porc_insc_muj']]
df_final_1[:1]
```
